**1.** Дан случайный массив, поменять знак у элементов, значения которых между 3 и 8

```python
arr = np.random.randint(0, 100, 10)
arr[arr.argmax()] = 0
print(arr)
```

**3.** Построить прямое произведение массивов (все комбинации с каждым элементом). На вход подается двумерный массив

```python
from itertools import product
arr = np.array([[1,2],[3,4]])
result = np.array(list(product(arr[0], arr[1])))
print(result)
```

**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
A = np.random.randint(0, 5, (8, 3))
B = np.random.randint(0, 5, (2, 2))
rows = []
for row in A:
    if all(any(item in row for item in b_row) for b_row in B):
        rows.append(row)
print(np.array(rows))
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
mat = np.random.randint(1, 4, (10, 3))
result = mat[~np.all(mat == mat[:, [0]], axis=1)]
print(result)
```

**6.** Дан двумерный массив. Удалить те строки, которые повторяются

```python
mat = np.random.randint(0, 3, (10, 2))
unique_rows = np.unique(mat, axis=0)
print(unique_rows)
```

* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
# without Numpy
def product_nonzero_diag_python(matrix):
    # matrix: list of lists
    m, n = len(matrix), len(matrix[0])
    prod = 1
    for i in range(min(m, n)):
        if matrix[i][i] != 0:
            prod *= matrix[i][i]
    return prod

X = [[1,0,1],[2,0,2],[3,0,3],[4,4,4]]
print(product_nonzero_diag_python(X))  # 3


# with Numpy
def product_nonzero_diag_numpy(X):
    diag = np.diag(X)
    return np.prod(diag[diag != 0])
```

* __Задача 2__: Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество.  
  Например, для x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) ответ True.

```python
# without Numpy
def same_multiset_python(x, y):
    return sorted(x) == sorted(y)

# with Numpy
def same_multiset_numpy(x, y):
    return np.array_equal(np.sort(x), np.sort(y))
```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит ноль. 
 Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ 5.

```python
# without Numpy
def max_after_zero_python(x):
    max_val = -np.inf
    for i in range(1, len(x)):
        if x[i-1] == 0 and x[i] > max_val:
            max_val = x[i]
    return max_val if max_val != -np.inf else None

# with Numpy
def max_after_zero_numpy(x):
    zero_positions = np.where(x[:-1] == 0)[0]
    if len(zero_positions) == 0:
        return None
    candidates = x[zero_positions + 1]
    return np.max(candidates)
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
# without Numpy
def rle_python(x):
    if len(x) == 0:
        return [], []
    values, counts = [], []
    cur_val = x[0]
    cnt = 1
    for i in range(1, len(x)):
        if x[i] == cur_val:
            cnt += 1
        else:
            values.append(cur_val)
            counts.append(cnt)
            cur_val = x[i]
            cnt = 1
    values.append(cur_val)
    counts.append(cnt)
    return values, counts

# with Numpy
def rle_numpy(x):
    if len(x) == 0:
        return np.array([]), np.array([])
    change = np.concatenate(([1], np.diff(x) != 0))
    values = x[change]
    counts = np.diff(np.concatenate((np.where(change)[0], [len(x)])))
    return values, counts
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
# without Numpy
import math
def euclidean_dist_matrix_python(X, Y):
    # X: list of vectors, Y: list of vectors
    dist = [[0]*len(Y) for _ in range(len(X))]
    for i, x in enumerate(X):
        for j, y in enumerate(Y):
            dist[i][j] = math.sqrt(sum((a-b)**2 for a,b in zip(x,y)))
    return dist

# with Numpy
def euclidean_dist_matrix_numpy(X, Y):
    # X: (m, d), Y: (n, d)
    X_sq = np.sum(X**2, axis=1).reshape(-1, 1)
    Y_sq = np.sum(Y**2, axis=1).reshape(1, -1)
    cross = X @ Y.T
    dist_sq = X_sq + Y_sq - 2 * cross
    return np.sqrt(dist_sq)

# speed comparison
from scipy.spatial.distance import cdist
import time

X = np.random.rand(500, 10)
Y = np.random.rand(600, 10)

t0 = time.time()
d1 = euclidean_dist_matrix_numpy(X, Y)
print("NumPy time:", time.time() - t0)

t0 = time.time()
d2 = cdist(X, Y)
print("scipy cdist time:", time.time() - t0)

print("Max difference:", np.max(np.abs(d1-d2)))
```
